**Extract movies with their actors and defining the role size for each actor.**

```python
moviesActorsDict={}
moviesYearsDict={}
moviesNames = movies_df.get("اسم الفيلم")
moviesActors=movies_df.get("تمثيل")
moviesYears=movies_df.get("تاريخ العرض")
for i in range(len(moviesNames)):
    temp1=split_string(moviesActors[i])
 
    temp2=actors_precentage(temp1)
    moviesActorsDict[moviesNames[i]+str(moviesYears[i])]=temp2;
    moviesYearsDict[moviesNames[i]+str(moviesYears[i])]=moviesYears[i]

```

**For each actor, gitting a list shows his role size in all movies he participated in**

```python

actDict={}

for movie,actor in moviesActorsDict.items():
       for key,value in actor.items():
            if key in actDict:
                actDict[key].append([value,moviesYearsDict[movie]])
            else:
                actDict[key]=[[value,moviesYearsDict[movie]]]
            
```

**Given an actor's name, plot a graph showing his performance through 1940s**

```python

def plot(actorName):
    figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    temp=actDict[actorName]
    l=len(temp)
    year=[]
    precentage=[]
    for i in range(l):
            precentage.append(temp[i][0])
            year.append(temp[i][1])
    Year =np.array(year)
    Actor_Role=np.array(precentage[::-1])
    plt.plot(Year, Actor_Role)
    plt.title('Actor average role/performace')
    plt.xlabel('Year').set_color('red')
    plt.ylabel('Actor Role').set_color('red')
    plt.show()
```



**Split a String based on Comma (،)**



```python
def split_string(string_to_split):
    crew_names = []
    try:
        for word in string_to_split.split(','):
            crew_names.append(word.strip())
    finally:
        return crew_names
```
