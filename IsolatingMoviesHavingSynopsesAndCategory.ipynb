```python
def isNaN(string):
    return string != string
```

```python
from csv import writer
def writeData(data,path):
    with open(path, 'a', encoding='UTF8', newline='') as f:
        w = writer(f)
        w.writerow(data)
        f.close()
```

```python
import math
for i in range(len(data)):
    if data.iloc[i]["ملخص"] !="ملخص" and not isNaN(data.iloc[i]["تصنيف الفيلم"]):
        writeData(data.iloc[i],"MoviesWithSynopsisAndCategory.csv")
    else:
        writeData(data.iloc[i],"MoviesWithoutSynopsisOrCategory.csv")
```
