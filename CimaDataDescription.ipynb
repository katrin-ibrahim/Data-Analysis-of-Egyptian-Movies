```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import arabic_reshaper as arabic_reshaper
from bidi.algorithm import get_display
```

```python
data=pd.read_csv(r'test.csv')
```

```python
def describe_column(colNames):
    
    fig, ax = plt.subplots()

    # hide axes
    fig.patch.set_visible(False)
    ax.axis('off')
    ax.axis('tight')

    plt.subplots_adjust(left=0.5, top=0.4)
    temp=[]
    for i in data[colNames].describe().values:
        if type(i)==str:
            temp.append(get_display(arabic_reshaper.reshape(i)))
        elif type(i)==np.float64:
            temp.append(round(i, 3))
        else:
            temp.append(i)
    df = pd.DataFrame(temp, columns=[get_display(arabic_reshaper.reshape(colNames))])

    ax.table(cellText=df.values, colLabels=df.columns,rowLabels=data[colNames].describe().index ,loc='center')

    fig.tight_layout()

    plt.show()
```

```python
def describe_dataframe(df):
    for colName in df.columns:
        describe_column(colName)
```

```python
describe_dataframe(data)
```
