```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import arabic_reshaper as arabic_reshaper
from bidi.algorithm import get_display
```

```python
df=pd.read_csv(r'test.csv')
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
def development_of_egyptian_categories_during_decades(df,cat):
    temp=[0,0,0,0]
    
    date=list(df.get("تاريخ العرض"))
    category=list(df.get("تصنيف الفيلم"))
    
    for i in range(len(category)):
        if cat==category[i]:
            if date[i]>=1940 and date[i]<1950 :
                temp[0]+=1
            elif  date[i]>=1950 and date[i]<1960 :
                temp[1]+=1
            elif  date[i]>=1960 and date[i]<1970 :
                temp[2]+=1
            elif  date[i]>=1970 and date[i]<1980 :
                temp[3]+=1

    decades=["1940s","1950s","1956s","1970s"]
    
    s = pd.Series(
        temp,
        index = decades
    )
    plt.figure(figsize=(16,5))
    
    addlabels(decades, temp)
    
    plt.title(get_display(arabic_reshaper.reshape(u'تطور '+cat+" المصرية خلال العقود")))
    plt.xlabel(get_display(arabic_reshaper.reshape("العقود")))
    plt.ylabel(get_display(arabic_reshaper.reshape("عدد الأفلام")))
    
    #plt.xticks(rotation=90)
    
    #Set tick colors:
    ax = plt.gca()
    ax.tick_params(axis='x', colors='red')
    ax.tick_params(axis='y', colors='red')
    
    s.plot( 
        kind='line', 
        color=list('c'),
        fontsize=15,
        linewidth=3
    )
    
    plt.show()

```

```python
development_of_egyptian_categories_during_decades(df,"دراما")
```

```python
development_of_egyptian_categories_during_decades(df,"رومانسي")
```
