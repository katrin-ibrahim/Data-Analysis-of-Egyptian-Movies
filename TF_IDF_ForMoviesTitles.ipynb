```python
import pandas as pd
```

```python
df= pd.read_excel(r'E:\DataAnalysis Internship\cima.xlsx')
```

```python
moviesTitles=df.get('اسم الفيلم')
moviesTitles
```

```python
doc=[]
s=" "
for i in range(len(moviesTitles)):
     doc.append(moviesTitles[i])
doc
```

```python
doc_splitted=[]
for i in range(len(doc)):
    doc_splitted.append(doc[i].replace('،','').split())
doc_splitted
```

```python
#word set contains all unique words through all movies
word_set=[]
for i in range(len(doc_splitted)):
    for j in range(len(doc_splitted[i])):
        if doc_splitted[i][j] not in word_set:
            word_set.append(doc_splitted[i][j])
```

```python
#count frequency of all words in specific document
def noOfOccurences(doc_splitted):
    wordDict=dict.fromkeys(word_set,0)
    for word in doc_splitted:
        wordDict[word]+=1
    return wordDict
```

```python
(noOfOccurences(doc_splitted[0]))
```

```python
#compute tf for this Document
def computeTF(Document):
    wordDictForTheDocument=noOfOccurences(Document)
    tfDict={}
    documentLength=len(Document)
    for word,count in wordDictForTheDocument.items():
        tfDict[word]=count/documentLength
        
    return  tfDict 
```

```python
tf = computeTF(doc_splitted[0])
tf
```

```python
import math
def computeIDF(Document):
    
    idfDict={}
    documentLength=len(Document)
    for i in range(len(Document)):
        wordDict=noOfOccurences(Document[i])
        for word,count in wordDict.items():
            if count>0:
                idfDict[word] =count+1
                
            
    for word,count in idfDict.items():  
        idfDict[word]=math.log(documentLength/(float(count)+1))
        
    return idfDict
```

```python
idf=computeIDF(doc_splitted)
idf
```

```python
def computeTFIDF(tf, idf):
  tfidf = {}
  for word, val in tf.items():
    tfidf[word] = val * idf[word]
  return tfidf
```

```python
tfidf = computeTFIDF(tf,idf)
tfidf
```
